# Categorical Variables

```python
# first, let's replace all blanks with NaN
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
# now return all categorical columns where over 50% of values are NaN

nulls_percent_df = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_df.columns = ['column_name', 'nulls_percentage']

columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>0.5]
display(columns_above_threshold['column_name'])
len(columns_above_threshold)
```

```python
categorical.isna().sum()
```

```python
drop_list = ['OSOURCE','ZIP'] # remove these as advised, they contain repeated info
drop_list = drop_list + list(columns_above_threshold['column_name'])
len(drop_list)
```

Here we will try to reduce the number of categories. An ideal way would have been to group the states into 
regions. But in this case we will group all the states with counts less than 2500 into one category "other"

```python
pd.DataFrame(categorical['STATE'].value_counts()).reset_index()
```

```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
categorical['STATE'].value_counts()
```

```python
categorical['MAILCODE'] = data['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
```

```python
categorical['MAILCODE'].value_counts()
```

```python
categorical['GENDER'].value_counts(dropna=False)
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
categorical['GENDER'] = np.where(categorical['GENDER'].isin(['U', 'J', 'C', 'A']),'other',categorical['GENDER'])
categorical['GENDER'].value_counts(dropna=False)
```

```python
# replace NaN with the modal value, then split column to reduce encoded columns later
display(categorical['DOMAIN'].value_counts(dropna=False))
categorical['DOMAIN'] = categorical['DOMAIN'].fillna('R2')
categorical['DOMAIN_A'] = list(map(lambda x: x[0], categorical['DOMAIN']))
categorical['DOMAIN_B'] = list(map(lambda x: x[1], categorical['DOMAIN']))
drop_list = drop_list + ['DOMAIN']
```

```python
# as discussed in lesson, these are too imbalanced to be of use
drop_list = drop_list + ['NOEXCH','MDMAUD','MDMAUD_R', 'MDMAUD_F','MDMAUD_A']
```

```python
# fill with median
categorical['CLUSTER'] = categorical['CLUSTER'].fillna('40')

# fill with unknown as there is already a category for such
categorical['HOMEOWNR'] = categorical['HOMEOWNR'].fillna('U')

# create a new category given the high number of NaN values
categorical['DATASRCE'] = categorical['DATASRCE'].fillna('0')

# fill with the mode
categorical['GEOCODE2'] = categorical['GEOCODE2'].fillna('A')
```

```python
# remove all RFA fields exept the two containing useful info
for col_name in categorical.columns:
    if "RFA" in col_name:
        drop_list.append(col_name)
drop_list.remove('RFA_2R')
drop_list.remove('RFA_2A')
```

```python
# this doesn't seem useful and contains many NaN values
drop_list = drop_list + ['AGEFLAG']
categorical.AGEFLAG.value_counts(dropna=False)
```

```python
#drop_list.remove('VETERANS')
#drop_list.remove('SOLIH')
categorical = categorical.drop(drop_list,axis=1)
```

```python
# first, remove all columns with more than 50% NaN values

nulls_percent_df1 = pd.DataFrame(numerical.isna().sum()/len(numerical)).reset_index()
nulls_percent_df1.columns = ['column_name', 'nulls_percentage']

columns_above_threshold1 = nulls_percent_df1[nulls_percent_df1['nulls_percentage']>0.5]

display(columns_above_threshold1['column_name'])
len(columns_above_threshold1)
```

```python
# as discussed in lesson, these do not contain useful info
drop_list =[]

for col_name in numerical.columns:
    if "ADATE" in col_name:
        drop_list.append(col_name)

drop_list = drop_list + list(columns_above_threshold1['column_name'])
```

```python
numerical = numerical.drop(drop_list, axis=1)
```

```python
# replace with the mean
numerical['AGE'] = numerical["AGE"].fillna(np.mean(numerical['AGE']))

# this is categorical, replace with the mode
numerical['INCOME'] = numerical['INCOME'].astype('object')
numerical['INCOME'] = numerical['INCOME'].fillna(5.0)

# this is categorical, replace with the mean
numerical['CLUSTER2'] = numerical['CLUSTER2'].fillna(np.ceil(np.mean(numerical['CLUSTER2'])))

# this is categorical, replace with the median
numerical['WEALTH2'] = numerical['WEALTH2'].astype('object')
numerical['WEALTH2'] = numerical['WEALTH2'].fillna('5.0')

# assume that this was the first donation, so no time difference between the two
numerical['TIMELAG'] = numerical['TIMELAG'].astype('object')
numerical['TIMELAG'] = numerical['TIMELAG'].fillna('0.0')
```
